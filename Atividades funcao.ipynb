1. Tabuada

```python
def tabuada(n):
  """
  Apresenta a tabuada do inteiro n,
  onde 1<= n <= 9
  """
  print(f"Tabuada do {n}")
  print(f"---------------")
  for mult in range(11):
    print(f" {mult} x {n} = {mult*n}")

tabuada(7)
```

3. n-esimo vertical

```python
def printNums(n):

  for i in range(n+1):
    for num in range(i):
      print(f"{num+1}", end=" ")
    print()

printNums(7)
```

5. Verificador de primo

```python
def verificadorDePrimo(n):
  for numero in range(1,n+1):
    if numero != 1 and numero != n:
      if n % numero == 0:
        return False
  return True

verificadorDePrimo(7)
```

7.  Implemente a função todos_iguais(sequencia) que retorna True se todos os elementos de sequencia forem iguais, e retorna False caso contrário

```python
def todosIguais(sequencia):
  if len(set(sequencia)) == 1:
    return True
  else:
    return False

todosIguais(["marcos", "marcos", "marcos", "marcos", "marcos"])
```

8. Implemente a função todos_diferentes(sequencia) que retorna True se todos os elementos de sequencia forem diferentes entre si, retorna False caso o contrário, isto é, pelo menos um seja igual a outro componente da sequencia

9. Implemente uma função que recebe como argumentos um número n e uma lista de números, indice_elemento(n, lista). A função retornará o índice da primeira ocorrência do número na lista e -1 se o número não estiver na lista

```python
def indiceElemento(n, lista):
  cont = 0
  for num in lista:
    if num == n:
      return cont
    cont+=1
  return -1

indiceElemento(15, [10, 2, 5, 4 , 3, 2, 15, 54, 3, 7])
```

10. Implemente uma função que recebe um número n e retorna a menor potência de 2 maior ou igual a n, por exemplo, pot2(14) retornará 16, pot2(42) retornará 64

```python
def pot2(num):
  resultado = 1
  while resultado < num:
    resultado *= 2
  return resultado

pot2(42)
```

11. Implemente uma função que dado um número natural maior do que 1, decomponha esse número em fatores primos. Por exemplo, se o argumento de entrada for 36, saída deverá ser [2, 2, 3, 3], porque 2 x 2 x 3 x 3 = 36

```python
def decompoePrimo(num):
  listaPrimos = []

  for i in range(2,num+1):
    while num%i == 0:
      listaPrimos.append(i)
      num /= i
  return listaPrimos

decompoePrimo(100)
```

12. Implemente a função maiorN(lista, N) que recebe uma lista de números quaisquer, um valor N, 1 <= N <= len(lista), e retorna o Nésimo maior valor na lista de números. Por exemplo, se n for 1, retorna o maior valor na lista, se n for 2, retorna o segundo maior valor na lista, etc.
Exemplo:

lista [5, -1, 7, 0, -3, 9]

N = 2

print(f'Em {lista) o {N}o. maior valor é maiorN (lista, N)}')

```python
lista = [5, -1, 7, 0, -3, 9]
def maiorN(lista, N):
  lista.sort(reverse=True)
  if N > len(lista):
    index = N - (len(lista)+1)
  else:
    index = N-1
  return lista[index], N

maior, N = maiorN(lista, 7)
print(f'Em {lista} o {N}°. maior valor é {maior}')
```
